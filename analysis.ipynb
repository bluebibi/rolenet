## 1. usa.gov data from bit.ly

- 이미 웹에 저장되어 있는 txt 파일을 URL로 접근하여 가져오기 위하여 urllib2 모듈 활용

```python
pathUrl = 'http://52.68.174.105/wp-content/uploads/2015/06/movies.csv'
```

```python
import urllib2
from pandas import read_csv
```

```python
raw_movie_data = urllib2.urlopen(pathUrl)
```

```python
frame = read_csv(raw_movie_data)
```

```python
frame[:]
```

```python
frame['spectators']
```

```python
diameter_counts = frame['diameter'].value_counts()
```

```python
diameter_counts
```

```python
%matplotlib inline
```

```python
diameter_counts.plot(kind='barh', rot=0)
```

```python
diameter_counts.plot(kind='line', rot=270)
```
